### Tipping Problem

Let’s create a fuzzy control system which models how you might choose to tip at a restaurant. When tipping, you consider the service and food quality, rated between 0 and 10. You use this to leave a tip of between 0 and 25%.

We would formulate this problem as:

#### Two input variables: 
   -  Service
       -  Universe (ie, crisp value range): How good was the service of the wait staff, on a scale of 0 to 10?
       -  Fuzzy set (ie, fuzzy value range): poor, average, good
   -  Food quality
       -  Universe: How tasty was the food, on a scale of 0 to 10?
       -  Fuzzy set: poor, average, good
       
#### One output variable: 
-  Tip
    -  Universe: How much should we tip, on a scale of 0% to 25%
    -  Fuzzy set: low, medium, high

#### Three fuzzy rules
-  IF the service was good or the food quality was good, THEN the tip will be high.
-  IF the service was average, THEN the tip will be medium.
-  IF the service was poor and the food quality was poor THEN the tip will be low.
   
#### Usage
-  If I tell this controller that I rated:
    -  the service as 9.8, and
    -  the quality as 6.5,
-  it would recommend I leave:
    -  a 19.85% tip.

### Fuzzy Inference System by skfuzzy

First, we need to import basic packages and define the input and output variables

```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
from skfuzzy import control as ctrl
%matplotlib inline

# New Antecedent/Consequent objects hold universe variables and membership
# functions
quality = ctrl.Antecedent(np.arange(0, 11, 1), 'quality')
service = ctrl.Antecedent(np.arange(0, 11, 1), 'service')
tip = ctrl.Consequent(np.arange(0, 26, 1), 'tip')
```

### Fuzzy Membership Functions

Here we can use the skfuzzy control system API to define the membership functions, where the function <font color=red>automf</font> allows to automatically set up membership functions for the variable. If you prefer to define specified membership functions one-by-one (for one input/output variable), <font color=red>fuzz.[type of membership]</font> can be used, for example, fuzz.trimf for triangular  membership fucntions. More options for the membership fucntion design can be found [here](http://pythonhosted.org/scikit-fuzzy/api/skfuzzy.membership.html#skfuzzy.membership.trimf). 

```python
# Auto-membership function population is possible with .automf(3, 5, or 7)
quality.automf(3)
service.automf(3)

# Custom membership functions can be built interactively with a familiar,
# Pythonic API
tip['low'] = fuzz.trimf(tip.universe, [0, 0, 13])
tip['medium'] = fuzz.trimf(tip.universe, [0, 13, 25])
tip['high'] = fuzz.trimf(tip.universe, [13, 25, 25])
```

### Fuzzy Rules

Now, we need to set up the following three simple rules :

 - If the food is poor OR the service is poor, then the tip will be low;
 - If the service is average, then the tip will be medium;
 - If the food is good OR the service is good, then the tip will be high.

```python
rule1 = ctrl.Rule(quality['poor'] | service['poor'], tip['low'])
rule2 = ctrl.Rule(service['average'], tip['medium'])
rule3 = ctrl.Rule(service['good'] | quality['good'], tip['high'])
```

### Using Fuzzy Inference System

After defining the fuzzy variables and rules, we can simply create a control system via:

```python
tipping_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
```

In order to simulate this control system, we will create a <font color=red>ControlSystemSimulation</font>. 

```python
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)
```

We can now simulate our control system by simply specifying the inputs and calling the <font color=red>computer</font> method. Suppose we rated the quality 6.5 out of 10 and the service 9.8 of 10.

```python
# Pass inputs to the ControlSystem using Antecedent labels with Pythonic API
# Note: if you like passing many inputs all at once, use .inputs(dict_of_data)
tipping.input['quality'] = 6.5
tipping.input['service'] = 9.8

# Crunch the numbers
tipping.compute()
```

Once computed, we can view the result as well as visualize it.

```python
print (tipping.output['tip'])
tip.view(sim=tipping)
plt.show()
```

The resulting suggested tip is 19.85%.

If you would like to put multiple samples (training or test) for evaluation, the following loop can be helpful. Suppose we have 3 inputs, with Quality=[6.5, 6.8,7.8] and Service=[9.2,9.8,10.5], how to get the associated **three outputs** (produced by the FIS)?

```python
import numpy as np
Quality=np.array([6.5, 6.8,8.8])
Service=np.array([9.2,9.8,10.2])
System_outputs=np.zeros(3,dtype=np.float64)
for i in range(3):
    tipping.input['quality'] = Quality[i]
    tipping.input['service'] = Service[i]
    tipping.compute()
    System_outputs[i]=tipping.output['tip']    ### Save each output in the object 'System_outputs'
print(System_outputs)  # should have three values/elements
```
