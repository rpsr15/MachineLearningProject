```python
import numpy as np
x1 = np.random.random((100,))
x2 = x1
x2[10] = 3
x2[50] = 4
x2[80] = 5
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(x2, 'o')
plt.show()
```

```python
#interquartile range is the rangle of middle 50% of the data
# how it helps to find the outlier 
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

Q1 = np.percentile(x2, 25)
Q3 = np.percentile(x2, 75)
d = Q3 - Q1
range = [Q1-1.5*d, Q3+1.5*d]
```

```python
import numpy as np
N = 100
input = np.arange(1,N+1 , dtype = float)/5
TargetOutput = 2+np.sin(input)
SystemOutput = TargetOutput + 0.15*np.random.random(input.shape)
RErr = np.sum(np.absolute(TargetOutput-SystemOutput)/np.absolute(TargetOutput))/N
print(RErr)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.plot(input, TargetOutput)
plt.plot(input, SystemOutput)
```
